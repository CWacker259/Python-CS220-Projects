```python
import csv

def process_csv(filename):
    example_file = open(filename, encoding="utf-8")
    example_reader = csv.reader(example_file)
    example_data = list(example_reader)
    example_file.close()
    return example_data

# Use process_csv to pull out the header and data rows
csv_rows = process_csv("airbnb.csv")
csv_header = csv_rows[0] # A list of the column headers
csv_data = csv_rows[1:] # The entire CSV data set besides the headers

def cell(row_idx, col_name):
    col_idx = csv_header.index(col_name)
    val = csv_data[row_idx][col_idx]
    if val == "":
        return None
    if type(val) == float:
        return float(val)
    if type(val) == int:
        return int(val)
    # optional: convert types based on column name, 
    # ensuring that the float and int values are appropriately converted
    return val
```

```python
def find_prices_within(lat_min, lat_max, long_min, long_max):
    prices = []
    for idx in range(len(csv_data)):
        if cell(idx, "latitude") != None:
            if cell(idx, "longitude") != None:
                if lat_min <= float(cell(idx, "latitude")) and float(cell(idx, "latitude")) <= lat_max:
                    if long_min <= float(cell(idx, "longitude")) and float(cell(idx, "longitude")) <= long_max:
                        prices.append(int(cell(idx, "price")))
    return prices
```

```python
def reviews_availability_ratio(neighborhood):
    reviews = []
    for idx in range(len(csv_data)):
        if cell(idx, "number_of_reviews") != None:
            if int(cell(idx, "availability_365")) != 0:
                if cell(idx, "availability_365") != None:
                    if cell(idx, "neighborhood") == neighborhood:
                        reviews.append(int(cell(idx, "number_of_reviews"))/int(cell(idx, "availability_365")))
    return sum(reviews)/len(reviews)
```
