```python
import csv

# Modified from https://automatetheboringstuff.com/chapter14/
def process_csv(filename):
    example_file = open(filename, encoding="utf-8")
    example_reader = csv.reader(example_file)
    example_data = list(example_reader)
    example_file.close()
    return example_data

vac_rows = process_csv("vaccinations.csv")
vac_header = vac_rows[0] # A list of the column headers
vac_data = vac_rows[1:] # The entire CSV data set besides the headers

cov_rows = process_csv("raw_covid_data.csv")
cov_header = cov_rows[0] # A list of the column headers
cov_data = cov_rows[1:] # The entire CSV data set besides the headers

def cell(row_idx, col_name):
    col_idx = vac_header.index(col_name)
    val = vac_data[row_idx][col_idx]
    if val == "":
        return None
    if col_name == "date":
        return val
    if col_name == "country":
        return val
    if col_name == "population":
        return int(val)
    if col_name == "daily_vaccinations":
        return int(val)
    if col_name == "total_vaccinations":
        return int(val)
    if col_name == "people_vaccinated":
        return int(val)
    if col_name == "people_fully_vaccinated":
        return int(val)
    # optional: convert types based on column name, 
    # ensuring that the float and int values are appropriately converted
    return val
```

```python
def daily_vaccinations_on(given_date):
    '''return a dictionary mapping each country to 
    the number of daily vaccinations on the given date; 
    if data is missing, the value for that country is None'''
    daily_vax_dict = {}
    for i in range(len(vac_data)):
        date = cell(i, "date")
        country =  cell(i, "country")
        daily_vax = cell(i, "daily_vaccinations")
        if date == given_date:
            daily_vax_dict[country] = daily_vax
            if daily_vax == None:
                continue
            else:
                daily_vax_dict[country] = daily_vax
    return daily_vax_dict
```

```python
import datetime

def get_number_of_days(start_date, end_date):
    day1 = datetime.datetime.strptime(start_date, '%m/%d/%Y') # The second argument is a format string to tell the function how to process the date string
    day2 = datetime.datetime.strptime(end_date, '%m/%d/%Y')
    delta = day2 - day1
    return delta.days

def fully_vaccinated_by(given_date):
    fully_vax_dict = {}
    for i in range(len(vac_data)):
        country = cell(i, "country")
        date = cell(i, "date")
        fully_vax = cell(i, "people_fully_vaccinated")
        if country not in fully_vax_dict:
            fully_vax_dict[country] = None
        if fully_vax == None:
            continue
        if get_number_of_days(date, given_date) >= 0:
            fully_vax_dict[country] = fully_vax
    return fully_vax_dict
```
