```python
def download(filename, url):
    if os.path.exists(filename):
        return (str(filename) + " already exists!")
    else:
        r = requests.get(url)
        r.raise_for_status()
        data = r.text
        f = open(filename, "w", encoding="utf-8")
        f.write(data)
        f.close()
    return (str(filename) + " created!")
```

```python
#Q12: Between the years 2019-2020 and 2021-2022, list the institutions which have seen an improvement in their "World Rank" by more than 500 ranks 
year_2019_ranking = rankings[rankings["Year"] == "2019-2020"][["World Rank", "Institution"]]
institutions_df = year_2021_ranking.merge(year_2019_ranking, how = 'inner', on = 'Institution') 
institutions_df["difference"] = institutions_df["World Rank_y"] - institutions_df["World Rank_x"] 
inst_improv = institutions_df[institutions_df["difference"] >= 500]
list(inst_improv["Institution"])
```

```python
#Q13: Between the years 2019-2020 and 2021-2022, which institution has the most significant change in its "World Rank"?
max_dif = institutions_df[institutions_df["difference"] == institutions_df["difference"].max()]["Institution"]
list(max_dif)[0] 
```

```python
#Q14: For all the three years, find the number of institutions which have a continuous improvement in their "World Rank".
year_2020_ranking = rankings[rankings["Year"] == "2020-2021"][["World Rank", "Institution"]]
institutions_df_1 = year_2020_ranking.merge(year_2019_ranking, how = 'inner', on = 'Institution')
institutions_df_2 = year_2021_ranking.merge(year_2020_ranking, how = 'inner', on = 'Institution')
institutions_df_1["difference"] = institutions_df_1["World Rank_y"] - institutions_df_1["World Rank_x"] 
institutions_df_2["difference"] = institutions_df_2["World Rank_y"] - institutions_df_2["World Rank_x"]  
pos_inst_1 = institutions_df_1[institutions_df_1["difference"] > 0]   
pos_inst_2 = institutions_df_2[institutions_df_2["difference"] > 0] 
pos_inst = pos_inst_2.merge(pos_inst_1, how = 'inner', on = 'Institution')  
both_improv = pos_inst[(pos_inst["difference_x"] > 0) & (pos_inst["difference_y"] > 0)] 
len(both_improv)
```
